```r
##########################################################################################################
# Taller TIN borrador
rm(list=ls()) # Para limpiar todas las variables
datos <- read.csv("https://raw.githubusercontent.com/JaFrese/DataFrames/main/geo_pol.csv",header = TRUE,sep = ";",dec = ",")
#View(datos)

# Ejercicio 1
# Transformar a header las filas (primera variable)
row.names(datos)=datos$pais
datos$pais=NULL

# Ejercicio 2
# Renombrar columnas
names(datos)=c('X1','X2','X3','X4','X5','X6','X7','X8','X9','X10')

# Ejercicio 3
# calcular componentes principales
pcgeopol <- prcomp(datos[,-1], scale=T)
```

```r
# Ejercicio 4
# analisis de componentes principales
names(pcgeopol)
pcgeopol$center # valores promedios de los componentes principales
pcgeopol$rotation # vectores propios colgados para cada componente (matriz gamma) / se diseña la ecuacion
pcgeopol$x # nueva tabla evaluado con los valores (sirve para crear los diagramas) / teorema espectral
pcgeopol$sdev # desviacion estandar de los componentes principales
pcgeopol$scale # valores promedios divididos por las desviaciones estandar (variables escaladas)
```

```r
# Ejercicio 5
# porcentajes de varianza explicada por cada componente	principal
pcgeopol.var <- (pcgeopol$sdev)^2
pve <- pcgeopol.var/sum(pcgeopol.var)
pve
plot(cumsum(pve), xlab="Componente Principal", ylab="Proporcion acumulada de varianza explicada", ylim=c(0,1),type='b')
```

```r

# Ejercicio 6
# grafico de barras componentes principales
library(ggplot2)
df <- data.frame(PC=c("PC1", "PC2", "PC3", "PC4", "PC5", "PC6", "PC7", "PC8", "PC9"),
                 pve=pve)
ggplot(data=df,aes(x=PC,y=pve,fill=PC))+
  geom_bar(stat = "identity", show.legend = F)+
  geom_text(aes(label=pve), position=position_dodge(width=0.9), vjust=-0.25,size=3)
```

```r
# Ejercicio 7
# grafico de torta/pastel/pie componentes principales
P <- ggplot(data=df,aes(x='',y=pve,fill=PC))+geom_bar(stat = "identity",width=2)
P <- P+geom_text(aes(y=pve,label = pve), position=position_dodge(width=0.9),
              vjust=-0.5,size=3)
P+coord_polar("y", start=0)
```

```r
# Ejercicio 8
# diagrama dispersión componentes principales
par(mfrow = c(1, 1))
y <- pcgeopol$x
par(mfrow=c(1,1))
plot(y[,c(1,2)],type="n",main="PC1 vs. PC2")
text(y[,c(1,2)],row.names(datos),cex=0.7)
plot(y[,c(3,2)],type="n",main="PC3 vs. PC2")
text(y[,c(3,2)],row.names(datos),cex=0.7)
plot(y[,c(1,3)],type="n",main="PC1 vs. PC3")
text(y[,c(1,3)],row.names(datos),cex=0.7)
plot(pcgeopol,main="")
par(mfrow=c(1,1))
```

```r
# Ejercicio 9
# diagrama biplot
biplot(pcgeopol)
```

```r
# Ejercicio 10
# reconocimiento	de	patrones separando	los	grupos	en	clusters
# determina la distancia entre las observaciones
pcgeopol.dist <- dist(y)
# aplica un método jerárquico para construir clusters:
pcgeopol.hclust <- hclust(pcgeopol.dist,method='ward.D')
# indica que usaremos 4 clusters
grupos.4 <- cutree(pcgeopol.hclust,4)
datos <- data.frame(datos,grupos.4=grupos.4)
# Graficamos PC1 vs. PC2
plot(y[, 1], y[, 2], pch =datos$grupos.4, col = datos$grupos.4,
     xlab = "PC1", ylab = "PC2", main = "Primero vs. Segundo PC",
     cex.lab = 1.2, cex.axis = 1.2, cex.main = 1.8,cex=1.5)
text(y[,c(1,2)],row.names(datos),cex=0.7)
```

```r
# Ejercicio 11
library(ggfortify)
df <- datos[,c(2:10)]
autoplot(prcomp(df,scale=T), data = datos,
         loadings = TRUE, loadings.colour = 'brown',
         loadings.label = TRUE, loadings.label.size = 4,label = TRUE,
         label.colour = grupos.4,shape = FALSE,label.size=3)
# correlacion las variables antiguas x1, x2, con los dos primeros componentes principales PC1, PC2
x <- df
(n = nrow(x))
(h=ncol(df))
# calculates eigenvalues and eigenvectors and sorts them by size
e <- eigen((n-1)*cov(x)/n)
e1 <- e$values/sum(e$values)
m <- apply(as.matrix(x),2,mean)
temp <- as.matrix(x-matrix(m,n,ncol(x),byrow=T))
r <- temp%*%e$vectors
# Calcula la correlacion entre los PC y las variables originales
r <- cor(cbind(r,x))
a=h+1
b=2*h
# La correlacion entre los dos PC mas importantes y las variables
r1 <- r[a:b,1:2]
```

```r
# grafico de las correlaciones entre los dos primeros PC y las variables originales
ucircle=cbind(cos((0:360)/180*pi),sin((0:360)/180*pi))
plot(ucircle,type="l",lty="solid",col="blue",xlab="Primer PC",ylab="Segundo PC",
     main="Análisis GeoPol",cex.lab=1.2,cex.axis=1.2,cex.main=1.8,lwd=2)
abline(h=0.0,v=0.0)
label=names(df)
text(r,label,cex=1.2)
```
