```r
#################### Clase 4 ##########################################
## Ejemplo 1 ##
# funcion de probabilidad

cdf <- function(x){
  ifelse(x>=0 &x<=20,1/20*x,0)
}
cdf(20)-cdf(15)
set.seed(123)
a <- sample(1:20,1000,replace = TRUE)
mean(a)

# varianza
varianza <- mean(a^2)-(mean(a))^2
varianza

# desviacion estandar
sqrt(varianza)

## Ejemplo 2 ##

df <- data.frame(x1=rnorm(2000),x2=rnorm(2000))
head(df)
library(ggplot2)
ggplot(df,aes(x = x1, y = x2)) + geom_point(color="darkblue")

round(cor(df$x1,df$x2))

# Matrices
A <- matrix(c(0,1,2,3,-1,0,4,1,1),nrow = 3,byrow = TRUE)
A
t(A) # A traspuesta
det(A) # determinante de A
dim(A) # dimension de A
A[1,3] # buscar valor por columna 1 y fila 3
solve(A) # inversa de la matriz A
diag(c(1,1)) # matriz de diagonal 1 y 1 (matriz 2x2)

B <- matrix(c(0,1,2,3,4,5,6,7,8),nrow = 3,byrow = TRUE)
B
A%*%B #producto matricial de A y B
round(A%*%solve(A)) # matriz A multiplicado por su matriz inversa

C <- matrix(c(13,-4,2,-4,13,-2,2,-2,10),nrow=3,byrow = TRUE)
C
l <- eigen(C) # entrega los valores y vectores propios de la matriz C
l$values # valores propios de la matriz C
l$vectors # vectores propios de la matriz C

# funcion de distribucion binomial
#install.packages("mvtnorm")
library(mvtnorm)
library(ggplot2)
set.seed(123)
n <- 2000
mu <- c(0,0)
ro <- 0.8
sig <- matrix(c(1,ro,ro,1),nrow = 2)
sig
n_random <- rmvnorm(n,mu,sig)
class(n_random) # tipo de variable
dim(n_random)
head(n_random,5)
df1 <- data.frame(x1=n_random[,1],x2=n_random[,2])
head(df1,5)

ggplot(df1,aes(x = x1, y = x2)) + geom_point(color="darkmagenta")
round(cor(df1$x1,df1$x2),1)

```

```r
################### Clase 5 ################################################
# Ejemplo 1.5: Matrix de covarianza/correlación para un dataset:
datos=read.table('https://raw.githubusercontent.com/JaFrese/DataFrames/main/bank2.dat')
names(datos)
head(datos)
names(datos)=c('X1','X2','X3','X4','X5','X6')
head(datos)
datos=data.frame(datos,label=c(rep('v',100),rep('f',100)))
head(datos)
cor(datos$X5,datos$X6)

#install.packages('ggcorrplot')
library(ggcorrplot)
names(datos)
X=datos[,-7]#eliminamos la última columna (label)
# Matriz de correlación
corr <- round(cor(X), 1)
corr
# Plot
ggcorrplot(corr,
           type = "lower",
           lab = TRUE,
           lab_size = 3,
           method="circle",
           colors = c("tomato2", "white", "springgreen3"),
           title="Correlograma",
           ggtheme=theme_bw)
# otra visualización de la correlación
#install.packages("corrplot")
library(corrplot)
corrplot.mixed(corr)
#Teorema espectral y propiedad de la matriz de covarianza
sigmaX=cov(X)
l=eigen(sigmaX)
l$values
P=l$vectors

round(t(P)%*%sigmaX%*%P,2)
X=as.matrix(X)
class(X)
dim(X)
round(cov(X%*%P),2)
```
